```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import joblib
from google.colab import files

# 1. Create a "Healthy" Baseline + "Anomalous" Spikes
# Features: [CPU_Percent, Memory_Percent, Latency_ms]
normal_data = np.random.normal(loc=[20, 30, 50], scale=[5, 5, 10], size=(100, 3))
anomalous_data = np.random.normal(loc=[90, 85, 800], scale=[5, 5, 50], size=(10, 3))

# Combine into one dataset
X = np.vstack([normal_data, anomalous_data])

# 2. Initialize the Model
# Contamination is the % of data we expect to be 'bad' (10/110 ≈ 0.09)
model = IsolationForest(contamination=0.1, random_state=42)
model.fit(X)

# 3. Save with our new naming convention
model_filename = 'aidp-anomaly-v1.joblib'
joblib.dump(model, model_filename)

# 4. Download to your local machine
files.download(model_filename)

print(f"Successfully exported {model_filename}!")
```
